Esto mostrará la página de manual para el comando `ls`.

#### Ejemplo de salida

```
LS(1)                     Manual de usuario de Linux                    LS(1)

NOMBRE
       ls - listar el contenido de un directorio

SINOPSIS
       ls [OPCIÓN]... [ARCHIVO]...

DESCRIPCIÓN
       Lista información sobre los ARCHIVOS (por defecto, los archivos en el directorio actual).
...
```

#### Salir del manual

Para salir de la página del manual y volver a la terminal, presiona la tecla `q`.

#### Otros ejemplos útiles

- Ver el manual de `cat`:
  ```bash
  man cat
  ```
